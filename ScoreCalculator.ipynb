Note1: Numbers are adjusted by the BigMac index, because why not. In effect, the more money you earn, the harder the negatives hit, and the less the positives contribute. Because it's easier for rich people to be green, and less excuse to not be green.

Note2: None of your data will be stored anywhere and the values you put in here cannot be seen by anyone.

Note3: This is semi scientific. Take it for what it is. The plugged numbers are somewhat roughly representative. The table is already populated with my best guess values for the global average.

```
Yearly_flying_trips   = 0.6        # How many single distance flights have you done the last 12 months?
Kg_meat_per_week      = 0.7        # How many kilograms of meat do you eat on average per week?
Ratio_of_red_meat     = 33         # Ratio of all meat you eat that is red, roughly (in percent)
Trees_planted_yearly  = 0.5        # How many trees have you planted (and seen take root) the last 12 months?
Trees_buried_yearly   = 0.001      # How many trees have you buried the last 12 months?
Money_used_on_stuff   = 71         # How much have you roughly used on buying stuff (not food) each month on average in US dollars ($)?
Distance_driven_ICE   = 400        # How many kilometers do you drive each month (with a fossil car), on average? For carpooling; divide by number of people
Money_invested        = 571        # How much money have you invested in green stocks/funds in US dollars ($)?
Money_contributed     = 5          # How much money do you contribute on average monthly to green charities, organizations, climate quotas, etc, in US dollars ($)?
Time_used             = 2          # How many hours do you use to fight the good fight, on average each month?

Bigmac_price          = 3.50      # What's the price of a BigMac where you live?
Hourly_wage           = 1.5        # Hourly wage, or hourly income if not salaried (roughly)

##########################################################################################################################################################
############################################################ HERE BE CALCULATIONS ########################################################################
##########################################################################################################################################################

IIBM = Hourly_wage/Bigmac_price # Your Bigmac index
flying = (Yearly_flying_trips*500*IIBM)/12
meat = (((Kg_meat_per_week*(Ratio_of_red_meat/100)*45) + (Kg_meat_per_week*(1-(Ratio_of_red_meat/100))*5))*IIBM)*4
trees = ((Trees_planted_yearly*500 + Trees_buried_yearly*1000)/IIBM)/12
consumerism = (Money_used_on_stuff/10)*IIBM
driving = (Distance_driven_ICE*0.13)*IIBM
time_money = (Money_invested/100 + Money_contributed + (Time_used*Hourly_wage)/10)/IIBM

print("######### Your monthly score: ###########")
print("# Score flying: {:>9.2f}               #".format(-flying))
print("# Score meat:   {:>9.2f}               #".format(-meat))
print("# Score trees:  {:>9.2f}               #".format(trees))
print("# Consumerism:  {:>9.2f}               #".format(-consumerism))
print("# Driving:      {:>9.2f}               #".format(-driving))
print("# Time & Money: {:>9.2f}               #".format(time_money))
print("#---------------------------------------#")
print("# Total score:  {:>9.2f}               #".format(time_money+trees-flying-meat-consumerism-driving))
print("#########################################")
print("## SHARE AND COMPETE, HIGHER IS BETTER ##")
```
