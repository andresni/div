```
Yearly_flying_trips   = 3        # How many single distance flights have you done the last 12 months?
Kg_meat_per_week      = 0.5      # How many kilograms of meat do you eat on average per week?
Ratio_of_red_meat     = 20       # Ratio of all meat you eat that is red, roughly (in percent)
Trees_planted_yearly  = 3        # How many trees have you planted (and seen take root) the last 12 months?
Trees_buried_yearly   = 0        # How many trees have you buried the last 12 months?
Money_used_on_stuff   = 5000     # How much have you roughly used on buying stuff each month on average?
Distance_driven_ICE   = 0        # How many kilometers do you drive each month (with a fossil car), on average? For carpooling; divide by number of people
Money_invested        = 40000    # How much money have you invested in green stocks/funds?
Money_contributed     = 500      # How much money do you contribute on average monthly to green charities, organizations, climate quotas, etc?
Time_used             = 50       # How many hours do you use to fight the good fight, on average each month?

Bigmac_price          = 110      # What's the price of a BigMac where you live?
Hourly_wage           = 200      # Hourly wage, or hourly income if not salaried (roughly)

##########################################################################################################################################################
############################################################ HERE BE CALCULATIONS ########################################################################
##########################################################################################################################################################

IIBM = Hourly_wage/Bigmac_price # Your Bigmac index
flying = Yearly_flying_trips*500*IIBM
meat = ((Kg_meat_per_week*(Ratio_of_red_meat/100)*45) + (Kg_meat_per_week*(1-(Ratio_of_red_meat/100))*5))*IIBM
trees = (Trees_planted_yearly*500 + Trees_buried_yearly*1000)/IIBM
consumerism = (Money_used_on_stuff/10)*IIBM
driving = (Distance_driven_ICE*300)*IIBM
time_money = (Money_invested/100 + Money_contributed + (Time_used*Hourly_wage)/10)/IIBM

print("########## Your scores: #################")
print("# Score flying: {:>9.2f}               #".format(-flying))
print("# Score meat:   {:>9.2f}               #".format(-meat))
print("# Score trees:  {:>9.2f}               #".format(trees))
print("# Consumerism:  {:>9.2f}               #".format(-consumerism))
print("# Driving:      {:>9.2f}               #".format(-driving))
print("# Time & Money: {:>9.2f}               #".format(time_money))
print("#---------------------------------------#")
print("# Total score:  {:>9.2f}               #".format(time_money+trees-flying-meat-consumerism-driving))
print("#########################################")
print("## SHARE AND COMPETE, HIGHER IS BETTER ##")
```
